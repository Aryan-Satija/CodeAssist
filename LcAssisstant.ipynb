```python
new_df = df.drop(columns = ['is_premium', 'solution_link', 'frequency', 'url', 'discuss_count', 'accepted', 'submissions', 'companies', 'likes', 'dislikes', 'rating', 'asked_by_faang', 'similar_questions'])
```

```python
new_df.head()
```

```python
new_df.info()
```

```python
def convert(x):
    l = []
    if(isinstance(x, str)):
        l = x.split(',')
    return l
new_df['related_topics'] = new_df['related_topics'].apply(convert)
```

```python
new_df.head()
```

```python
new_df.info()
```

```python
new_df['description'] = new_df['description'].apply(lambda x : x.split())
```

```python
new_df.head()
```

```python
new_df.sample(5)
```

```python
new_df['difficulty'] = new_df['difficulty'].apply(lambda x : [x])
```

```python
new_df['acceptance_rate'] = new_df['acceptance_rate'].apply(lambda x : [x])
```

```python
new_df['acceptance_rate'] = new_df['acceptance_rate'].apply(lambda x : [  str(int(round(x[0], -1)))  ])
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['description'] + new_df['difficulty'] + new_df['acceptance_rate'] + new_df['related_topics']
```

```python
new_df.head()
```

```python
new_df.drop(columns = ['description', 'difficulty', 'acceptance_rate', 'related_topics'], inplace = True)
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem('loving')
```

```python
ps.stem('loved')
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_df['tags'] = new_df['tags'].apply(stem)
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
new_df.loc[[1]]
```

```python
sorted(list(enumerate(similarity[3])), key = lambda x : x[1], reverse = True)[1 : 6]
```

```python
def recommend(problem):
    ind = new_df[new_df['title'] == problem].index[0]
    problem_inds = sorted(list(enumerate(similarity[ind])), key = lambda x : x[1], reverse = True)[1 : 6]
    for prob_ind in problem_inds:
        print(new_df.iloc[prob_ind[0]].title)
```

```python
def recommend(problem_id):
    ind = df[df['id'] == problem_id].index[0]
    recommended_indices = sorted(list(enumerate(similarity[ind])), key = lambda x : x[1], reverse = True)[1: 6]
    response = []
    for prob_ind in recommended_indices:
        x = df.loc[prob_ind[0]]
        response.append({
            "id": x.iloc[0],
            "name": x.iloc[1],
            "description": x.iloc[2],
            "is_premium": x.iloc[3],
            "difficulty": x.iloc[4],
            "acceptance_rate": x.iloc[6],
            "url": x.iloc[8],
            "accepted": x.iloc[10],
            "submissions": x.iloc[11],
            "companies": x.iloc[12].split(',') if isinstance(x.iloc[12], str) else [],
            "related_topics": x.iloc[13].split(',') if isinstance(x.iloc[13], str) else [],
            "likes": x.iloc[14],
            "dislikes": x.iloc[15],
            "asked_by_faang": x.iloc[17],
        })        
    for x in response:
        print(x)
```
